```python
import pandas as pd 
import numpy as  np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"Downloads/dev.csv")
df.head()
```

```python
#droping the columns we dont need 
columns=df.columns.values.tolist()
for i in df.columns.values:
    year = list(i)
    try:
        year=int(''.join(year[len(year)-5:len(year)-1]))
        if year<2021 and year>1990:
            df=df.drop(i,axis=1)
    except Exception as e:
        pass
df=df.drop(['ISO3','Human Development Groups','UNDP Developing Regions','HDI Rank (2021)'],axis=1)
df2=df.query("Country == 'Algeria' | Country == 'Egypt' | Country == 'Morocco' | Country == 'Tunisia' | Country == 'Canada'")
df2
```

```python
df2.isnull()
```

```python
### Human Development Index (1990)
columns=df2.columns.values.tolist()
i=1
while(i<len(columns)-1):
    plt.figure(figsize=(8,4))
    sns.barplot(x = df2['Country'], y = df2[columns[i]], data = df2)
    sns.barplot(x = df2['Country'], y = df2[columns[i+1]], data = df2, hue = df2[columns[i]], palette=["#c4f8ef","#ffbdd6","#bde0ff"])
    plt.legend([],[], frameon=False)
    print(columns[i]+" and (2020)")
    plt.show()
    i=i+2
```
